```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

```

```python
imp = SimpleImputer()
data_test = pd.read_csv("/mnt/c/Users/ADMIN/Desktop/NUCLEI - TITANIC/test.csv")
data_train = pd.read_csv("/mnt/c/Users/ADMIN/Desktop/NUCLEI - TITANIC/train.csv")
data_train = data_train[data_train['Embarked'].notna()]
data_train['Embarked'].isnull().sum()
data_train.describe()
```

```python
new_data_train = data_train.drop(['PassengerId','Name','Cabin','Ticket','Survived'], axis=1)
new_data_train.head()
```

```python
new_data_test = data_test.drop(['PassengerId','Name','Cabin','Ticket'], axis=1)
new_data_test.head()
new_data_test['Embarked'].isnull().sum()
```

```python
sex = {'male': 1,'female': 2}
emb = {'S': 1,'C': 2, 'Q': 3, None:0}
new_data_train.Sex = [sex[item] for item in new_data_train.Sex]
new_data_test.Sex = [sex[item] for item in new_data_test.Sex]
new_data_train.Embarked = [emb[item] for item in new_data_train.Embarked]
new_data_test.Embarked = [emb[item] for item in new_data_test.Embarked]
new_data_train.head()
```

```python
y = data_train["Survived"]
X = imp.fit_transform(new_data_train)
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
X_test = imp.fit_transform(new_data_test)
A = np.zeros((10,10))

demo_model = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=1, criterion='entropy')
demo_model.fit(train_X, train_y)
val_predict = demo_model.predict(val_X)
val_mae = mean_absolute_error(val_predict, val_y)
print(val_mae)
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=1)
model.fit(X, y)
```

```python
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': data_test.PassengerId, 'Survived': predictions})
output.to_csv('final_predict.csv', index=False)
output
```
